# Q.7) Reverse tuple and remove element, combine with another

```python
tup = (1, 2, 3, 4, 5)
element_to_remove = 3

reversed_tup = tuple(reversed(tup))
print(f"Reversed: {reversed_tup}")

removed_tup = tuple(x for x in reversed_tup if x != element_to_remove)
print(f"After removing {element_to_remove}: {removed_tup}")

another_tup = (10, 20, 30)
combined = removed_tup + another_tup
print(f"Combined: {combined}")
```
