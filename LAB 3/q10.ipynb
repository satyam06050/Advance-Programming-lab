# Q.10) Find longest increasing continuous subsequence

```python
lst = [1, 3, 4, 4, 5, 6, 7, 3, 4, 5, 6]

max_length = 0
max_subseq = []
current_subseq = [lst[0]]

for i in range(1, len(lst)):
    if lst[i] > lst[i-1]:
        current_subseq.append(lst[i])
    else:
        if len(current_subseq) > max_length:
            max_length = len(current_subseq)
            max_subseq = current_subseq[:]
        current_subseq = [lst[i]]

if len(current_subseq) > max_length:
    max_length = len(current_subseq)
    max_subseq = current_subseq[:]

print(f"List: {lst}")
print(f"Longest increasing subsequence: {max_subseq}")
print(f"Length: {max_length}")
```
