```python
import time
import random
import numpy as np

# crrate matrix
A_list = []
B_list = []

#  matrix A
for i in range(100):
    row = []
    for j in range(100):
        row.append(random.randint(1, 10))
    A_list.append(row)

# matrix B
for i in range(100):
    row = []
    for j in range(100):
        row.append(random.randint(1, 10))
    B_list.append(row)

# Result matrix initialized with 0
C_list = []
for i in range(100):
    row = []
    for j in range(100):
        row.append(0)
    C_list.append(row)


start_time_list = time.time()

# Multiply matrices using normal Python loops
for i in range(100):
    for j in range(100):
        for k in range(100):
            C_list[i][j] += A_list[i][k] * B_list[k][j]

end_time_list = time.time()

print("Time taken using Python lists:",
      end_time_list - start_time_list, "seconds")


# -----------------------------
# MATRIX MULTIPLICATION USING NUMPY
# -----------------------------

# Create two 100x100 matrices using NumPy
A_np = np.zeros((100, 100), dtype=int)
B_np = np.zeros((100, 100), dtype=int)

for i in range(100):
    for j in range(100):
        A_np[i][j] = np.random.randint(1, 10)
        B_np[i][j] = np.random.randint(1, 10)


start_time_np = time.time()

# Multiply matrices using NumPy
C_np = np.dot(A_np, B_np)

end_time_np = time.time()

print("Time taken using NumPy:",
      end_time_np - start_time_np, "seconds")
```
