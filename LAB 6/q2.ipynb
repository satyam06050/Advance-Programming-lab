# Q.2) 2D NumPy array operations

```python
import numpy as np

arr = np.random.randint(1, 101, (10, 5))
print("Array:")
print(arr)

print(f"\nSum of odd numbers: {np.sum(arr[arr % 2 != 0])}")
print(f"Sum of even numbers: {np.sum(arr[arr % 2 == 0])}")
print(f"Sum of diagonal elements: {np.sum(np.diag(arr[:5, :5]))}")
print(f"\nTranspose:\n{arr.T}")
print(f"\nRows 2-6:\n{arr[1:6]}")

arr_modified = arr.copy()
arr_modified[1:6] *= -1
print(f"\nAfter multiplying rows 2-6 by -1:\n{arr_modified}")

flattened = arr[1:6].flatten()
print(f"\nFlattened rows 2-6: {flattened}")

print(f"Sum of all elements: {np.sum(arr)}")
print(f"Mean: {np.mean(arr)}")
print(f"Median: {np.median(arr)}")
print(f"Standard Deviation: {np.std(arr)}")

sorted_flat = np.sort(arr.flatten())
print(f"Flattened and sorted: {sorted_flat}")
```
