# Q.4) Solve systems of linear equations using inverse matrix method

```python
import numpy as np

print("i) x + 2y = -4, 2x + 4y = -8")
A1 = np.array([[1, 2], [2, 4]], dtype=float)
B1 = np.array([-4, -8], dtype=float)
det1 = np.linalg.det(A1)
print(f"Determinant: {det1}")
if det1 != 0:
    X1 = np.linalg.solve(A1, B1)
    print(f"Solution: x = {X1[0]}, y = {X1[1]}")
else:
    print("No unique solution (singular matrix)")

print("\nii) x + 2y + 3z = 6, 2x - 3y + 2z = 14, 3x + y - z = -2")
A2 = np.array([[1, 2, 3], [2, -3, 2], [3, 1, -1]], dtype=float)
B2 = np.array([6, 14, -2], dtype=float)
X2 = np.linalg.solve(A2, B2)
print(f"Solution: x = {X2[0]}, y = {X2[1]}, z = {X2[2]}")
```
