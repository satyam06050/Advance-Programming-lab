# Q.14) Convert decimal to binary, octal, hexadecimal using switch case

```python
decimal = int(input("Enter a decimal number: "))

print("Choose conversion:")
print("1) Binary")
print("2) Octal")
print("3) Hexadecimal")

choice = int(input("Enter your choice (1-3): "))

# Switch case using match-case (Python 3.10+) or if-elif
if choice == 1:
    result = bin(decimal)[2:]  # Remove '0b' prefix
    print(f"Binary: {result}")
elif choice == 2:
    result = oct(decimal)[2:]  # Remove '0o' prefix
    print(f"Octal: {result}")
elif choice == 3:
    result = hex(decimal)[2:]  # Remove '0x' prefix
    print(f"Hexadecimal: {result}")
else:
    print("Invalid choice!")
```
