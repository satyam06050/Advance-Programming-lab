# Q.14) Convert decimal to binary, octal, hexadecimal using switch case

```python
decimal = int(input("Enter a decimal number: "))

print("Choose conversion:")
print("1) Binary")
print("2) Octal")
print("3) Hexadecimal")

choice = int(input("Enter your choice (1-3): "))

if choice == 1:
    n = decimal
    result = ""

    if n == 0:
        print("Binary: 0")
    else:
        while n > 0:
            rem = n % 2
            result += str(rem)
            n = n // 2

        binary = result[::-1]
        print("Binary:", binary)

elif choice == 2:
    pass

elif choice == 3:
   pass
else:
    print("Invalid choice!")
```
