# Pattern Programs

```python
# Star Triangle Pattern
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print('*' * i)
```

```python
# Number Triangle Pattern
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# Pyramid Pattern
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```
