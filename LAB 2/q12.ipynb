# Q.12) Generate Fibonacci series up to specified terms

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1
print("Fibonacci series:")

for i in range(n):
    if i == 0:
        print(a, end=" ")
    elif i == 1:
        print(b, end=" ")
    else:
        c = a + b
        print(c, end=" ")
        a, b = b, c

print()  # New line
```
