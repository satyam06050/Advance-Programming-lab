```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
profit=df["total_profit"]
months=df["month_number"]

plt.plot(months,profit)

plt.title('Company profit per month')
plt.xlabel("months")
plt.ylabel("profits")
plt.show()
```

```python
plt.plot(months, profit,
         linestyle="--",
         color="r",
         linewidth=3,
         label="Monthly Profit")

plt.xlabel("Months")
plt.ylabel("Profit")
plt.title("Company Profit per Month")

plt.legend(loc='lower right')
plt.show()
```

```python
months=df["month_number"]

plt.plot(months, df["facecream"], label="Face Cream",marker='o')
plt.plot(months, df["facewash"], label="Face Wash",marker='o')
plt.plot(months, df["toothpaste"], label="Toothpaste",marker='o')
plt.plot(months, df["bathingsoap"], label="Bathing Soap",marker='o')
plt.plot(months, df["shampoo"], label="Shampoo",marker='o')
plt.plot(months, df["moisturizer"], label="Moisturizer",marker='o')

plt.xlabel("Months")
plt.ylabel("Sales / Profit")
plt.title("Company Sales Data")

plt.legend(loc="upper left")   
plt.show()
```

```python
 #4   Read toothpaste sales data of each month and show it using a scatter plot

months=df["month_number"]

plt.scatter(months, df["toothpaste"], color="g", label="Toothpaste Sales")
plt.grid(True,linestyle="--")
plt.legend(loc= 'upper left')
```
