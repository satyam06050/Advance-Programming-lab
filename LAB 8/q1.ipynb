```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
months = df['month_number']
```

# Exercise 1
Read Total profit of all months and show it using a line plot

```python
profit = df['total_profit']
plt.plot(months, profit)
plt.xlabel('Month number')
plt.ylabel('Total Profit')


plt.title('Company profit per month')
plt.show()
```

# Exercise 2
Get total profit of all months and show line plot with the following Style properties

```python
profit = df['total_profit']
plt.plot(months, profit, linestyle='--', color='red', marker='o', linewidth=3, label='Profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.show()
```

# Exercise 3
Read all product sales data and show it using a multiline plot

```python
plt.plot(months, df['facecream'], label='Face cream', marker='o')
plt.plot(months, df['facewash'], label='Face Wash', marker='o')
plt.plot(months, df['toothpaste'], label='ToothPaste', marker='o')
plt.plot(months, df['bathingsoap'], label='Bathing Soap', marker='o')
plt.plot(months, df['shampoo'], label='Shampoo', marker='o')
plt.plot(months, df['moisturizer'], label='Moisturizer', marker='o')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.show()
```

# Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(months, df['toothpaste'], label='Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title('Tooth paste Sales data')
plt.grid(True, linestyle='--')
plt.show()
```

# Exercise 5
Read face cream and facewash product sales data and show it using the bar chart

```python
plt.bar([a-0.25 for a in months], df['facecream'], width=0.25, label='Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in months], df['facewash'], width=-0.25, label='Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Facewash and facecream sales data')
plt.grid(True, linewidth=1, linestyle="--")
plt.show()
```

# Exercise 6
Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
plt.bar(months, df['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('bathingsoap sales data')
plt.grid(True, linewidth=1, linestyle="--")
plt.savefig('bathingsoap_sales_data.png')
plt.show()
```

# Exercise 7
Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
profit = df['total_profit']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profit, profit_range, label='Profit data')
plt.xlabel('profit range')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.title('Profit data')
plt.show()
```

# Exercise 8
Calculate total sale data for last year for each product and show it using a Pie chart

```python
sales_data = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), 
              df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(sales_data, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

# Exercise 9
Read Bathing soap facewash of all months and display it using the Subplot

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(months, df['bathingsoap'], label='Bathing Soap Sales Data', color='black', marker='o', linewidth=3)
axarr[0].set_title('Sales data of a Bathingsoap')
axarr[1].plot(months, df['facewash'], label='Face Wash Sales Data', color='red', marker='o', linewidth=3)
axarr[1].set_title('Sales data of a facewash')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

# Exercise 10
Read all product sales data and show it using the stack plot

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(months, df['facecream'], df['facewash'], df['toothpaste'], 
              df['bathingsoap'], df['shampoo'], df['moisturizer'], 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
