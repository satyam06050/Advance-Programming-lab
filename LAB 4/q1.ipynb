# Q.1) Create set, add elements, set operations

```python
set1 = {1, 2, 3, 4, 5}
set1.add(6)
set1.add(3)
set1.add(4)
print(f"Set1: {set1}")

set2 = {4, 5, 6, 7, 8}
print(f"Set2: {set2}")

print(f"Union: {set1.union(set2)}")
print(f"Intersection: {set1.intersection(set2)}")
print(f"Difference: {set1.difference(set2)}")
```
