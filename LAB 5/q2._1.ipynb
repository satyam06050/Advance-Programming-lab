# Q.2) Employee Hierarchy with OOP

```python
class Employee:
    companyName = "Tech Corp"
    
    def __init__(self, name, age, experience, department, salary):
        self.name = name
        self.__age = age
        self.experience = experience
        self.department = department
        self.__salary = salary
    
    def display(self):
        print(f"Name: {self.name}, Age: {self.__age}, Experience: {self.experience}, Dept: {self.department}")


class Manager(Employee):
    def __init__(self, name, age, experience, department, salary):
        super().__init__(name, age, experience, department, salary)
        self.teams = []
    
    def add_team(self, team):
        self.teams.append(team)
    
    def total_teams(self):
        return len(self.teams)


class Developer(Employee):
    def __init__(self, name, age, experience, department, salary, language):
        super().__init__(name, age, experience, department, salary)
        self.language = language
    
    def display(self):
        super().display()
        print(f"Languages: {self.language}")


emp = Employee("Raj", 25, 2, "IT", 50000)
emp.display()

mgr = Manager("Priya", 35, 8, "IT", 80000)
mgr.add_team("Team A")
mgr.add_team("Team B")
print(f"Manager: {mgr.name}, Teams: {mgr.total_teams()}")

dev = Developer("Arun", 28, 4, "IT", 60000, "Python, Java")
dev.display()
```
