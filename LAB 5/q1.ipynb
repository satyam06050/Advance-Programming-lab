```python
# Define the Box class
class Box:
    def __init__(self, length, width, height):
        
        self.length = length
        self.width = width
        self.height = height
    
    def volume(self):
        
        return self.length * self.width * self.height


# Create an object of the Box class
box1 = Box(5, 4, 3)

# Print the volume
print(f"Length: {box1.length}")
print(f"Width: {box1.width}")
print(f"Height: {box1.height}")
print(f"Volume of the box: {box1.volume()} cubic units")
```
