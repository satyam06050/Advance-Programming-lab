# Q.2) Reverse a 3-digit number using operators only

```python
num = int(input("Enter a 3-digit number: "))

# Extract digits using operators
digit1 = num % 10          # Last digit
digit2 = (num // 10) % 10  # Middle digit  
digit3 = num // 100        # First digit

# Form reverse number
reverse = digit1 * 100 + digit2 * 10 + digit3

print(f"Original number: {num}")
print(f"Reversed number: {reverse}")
```
