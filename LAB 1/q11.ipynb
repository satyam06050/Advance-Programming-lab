# Q.11) Check if 3-digit number is palindrome without using loop

```python
num = int(input("Enter a 3-digit number: "))

# Extract digits
digit1 = num % 10          # Last digit
digit2 = (num // 10) % 10  # Middle digit
digit3 = num // 100        # First digit

# Check palindrome condition
is_palindrome = digit1 == digit3

print(f"Number: {num}")
print(f"Is palindrome: {is_palindrome}")
```
