# Q.4) Check if element exists in set using operator only

```python
# Initialize a set
my_set = {1, 2, 3, 4, 5, "apple", "banana"}

# Element to check
element = input("Enter element to check: ")

# Convert to int if it's a digit
element = int(element) if element.isdigit() else element

# Check using 'in' operator
result = element in my_set

print(f"Set: {my_set}")
print(f"Element '{element}' is in set: {result}")
```
