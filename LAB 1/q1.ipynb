# Q.1) Variables, types, deletion and comparison

```python
# Assign different types of values
name = "Alice"
age = 25
height = 5.8
is_student = True
marks = [85, 90, 78]

# Print all variables
print("Variables:")
print(f"name: {name}")
print(f"age: {age}")
print(f"height: {height}")
print(f"is_student: {is_student}")
print(f"marks: {marks}")
```

```python
# Display types of all variables
print("Variable types:")
print(f"type(name): {type(name)}")
print(f"type(age): {type(age)}")
print(f"type(height): {type(height)}")
print(f"type(is_student): {type(is_student)}")
print(f"type(marks): {type(marks)}")
```

```python
# Delete a particular variable
del height
print("After deleting 'height' variable:")
print("Available variables:", [var for var in locals() if not var.startswith('_')])
```

```python
# Assign two integer variables and compare
num1 = 15
num2 = 20
print(f"num1 = {num1}, num2 = {num2}")
print(f"num1 == num2: {num1 == num2}")
print(f"num1 > num2: {num1 > num2}")
print(f"num1 < num2: {num1 < num2}")
```
